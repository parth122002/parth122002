```python
import pandas as pd
```

```python
df=pd.read_excel('covtype.csv')
```

```python
df
```

```python
del df['Unnamed: 10']
df.head(3)
```

```python
df.drop('Serial_No',inplace=True,axis=1)
```

```python
y=df['Research']
y.head(3)
```

```python
x=df.drop('Research',inplace=False,axis=1)
x.head(3)
```

```python

from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression().fit(x,y)
classifier.score(x,y)
```

```python
from sklearn.preprocessing import Normalizer
```

```python
normalizer = Normalizer().fit(x)
norm_x = normalizer.transform(x)
norm_x
```

```python
classifier = LogisticRegression().fit(norm_x,y)
classifier.score(norm_x,y)
```
